```python
#Q2
def my_max(x):
    max = x[0]
    for i in x:
        if i > max:
            max = i
    return max
```

```python
x = [7, 9, 10, 5, 8, 3, 4, 6, 2, 1]

print(my_max(x))
```

```python
#Q3

x = [7, 9, 10, 5, 8, 3, 4, 6, 2, 1]

def my_n_max(x, n):
    # write your function code here
    out = []
    count = 0
    while count < n:
        maximum = my_max(x)
        out.append(maximum)
        x.remove(maximum)
        count += 1
    return out
```

```python
# Output = [10, 9, 8]
out = my_n_max(x, 3)
print(out)
```

```python
#Q4
import math 
def my_trig_odD_even(m):
    q = []
    for i in range(len(m)):
        for j in range(len(m[i])):
            if (i+j) % 2 == 0:
                q.append(math.sin(m[i][j]))
            elif (i+j)%2 == 1:
                q.append(math.cos(m[i][j]))
    return q
```

```python
m = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(my_trig_odD_even(m))
```

```python
# Q5
import numpy as np

def my_mat_mult(p, q):
    m = [[0 for _ in range(len(q[0]))] for _ in range(len(p))]
    # write your function code here
    for i in range(len(p)):
        for j in range(len(q[0])):
            for k in range(len(p[0])):
                m[i][j] += p[i][k] * q[k][j]
    return m
```

```python
P = np.ones((3, 3))
my_mat_mult(P, P)
```

```python
P = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
Q = np.array([[1, 1, 1], [2, 2, 2], [3, 3, 3], [4, 4, 4]])
my_mat_mult(P, Q)
```

```python
#Q6
def my_saving_plan(P0, i, goal):
    # write your function code here
    total = P0
    years = 0

    while total <= goal:
        total = (1+i) * total
        years += 1
    return years
```

```python
my_saving_plan(1000, 0.05, 2000)
```

```python
my_saving_plan(1000, 0.07, 2000)
```

```python
my_saving_plan(500, 0.07, 2000)
```

```python
#Q7

def my_find(M):
    out = []
    for i in range(len(M)):
        if M[i] == 1:
            out.append(i)
    return out
```

```python
M = [1, 0, 1, 1, 0]
my_find(M)
```

```python
#Q8

def my_monopoly_dice():
    throw1 = int(input())
    throw2 = int(input())
    sum = 0
    if throw1 != throw2:
        sum = throw1 + throw2
    elif throw1 == throw2:
        sum = throw1 + throw2
        print("Another die throw!")
        sum += my_monopoly_dice()
    return sum
```

```python
#Q9

def is_prime(n):
    if n <= 1:
        return 0
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return 0
    return 1
```

```python
is_prime(11)
```

```python
#Q10

def my_n_primes(n):
    p = []
    for i in range(n+1):
        if is_prime(i) == 1:
            p.append(i)
    return p
```

```python
my_n_primes(10)
```

```python
#Q11

def iter_fib(n):
    fib = np.ones(n)
    for i in range(2, n):
        fib[i] = fib[i - 1] + fib[i - 2]
    return fib
    
def my_n_fib_primes(n):
    # write your function code here
    x = iter_fib(n)
    fib_primes = []
    for i in x:
        if is_prime(i) == 1:
            fib_primes.append(int(i))
    return fib_primes
```

```python
my_n_fib_primes(8)
```

```python
my_n_fib_primes(30)
```

```python
#Q12

def my_trig_odd_even(M):
    # write your function code here
    Q = [[0 for _ in range(len(M[0]))] for _ in range(len(M))]
    for i in range(len(M)):
        for j in range(len(M[i])):
            if M[i][j] % 2 == 0:
                Q[i][j] = math.cos(math.pi / M[i][j])
            elif M[i][j] % 2 == 1:
                Q[i][j] = math.sin(math.pi / M[i][j])
    return Q
```

```python
M = [[3, 4], [6, 7]]
my_trig_odd_even(M)
```

```python
# Q13
def my_connectivity_mat_2_dict(C, names):
    # write your function code here
    node = {}
    C1=[[0 for _ in range(len(C[0]))] for _ in range(len(C))]
    for i in range(len(C)):
        for j in range(len(C[i])):
            if C[i][j] == 1:
                C1[i][j] = j + 1
    #C1.remove(0)
    for i in range(len(C1)):
        C1[i] = [x for x in C1[i] if x != 0]
    for k, v in zip(names, C1):
        node[k] = v
    return node
```

```python
C = [[0, 1, 0, 1], [1, 0, 0, 1], [0, 0, 0, 1], [1, 1, 1, 0]]
names = ['Los Angeles', 'New York', 'Miami', 'Dallas']
```

```python
my_connectivity_mat_2_dict(C, names)
```
